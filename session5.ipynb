```python
# A manufacturer claims the average battery life of its phones is 300 hours. Test the claim at α = 0.05.
# A consumer group tests 50 phones and finds the average battery life is 290 hours with a sample standard deviation of 40 hours.
# What's available:
# mean_pop = 300 hours
# n= 50
# mean_sample = 290 hours
# sd_sample= 40
# alpha= 0.05
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Step 1: Define the known values
mu_0 = 300      # Claimed population mean (hypothesized)
x_bar = 290     # Sample mean from tested phones
s = 40          # Sample standard deviation
n = 50          # Sample size
alpha = 0.05    # Significance level for 95% confidence

# Step 2: Calculate the Standard Error (SE)
se = s / np.sqrt(n)

# Step 3: Calculate the t-statistic
t_stat = (x_bar - mu_0) / se

# Step 4: Degrees of freedom
df = n - 1

# Step 5: Calculate the two-tailed p-value
p_val_two_tail = 2 * t.cdf(t_stat, df=df)

# Step 6: Find the critical t-value for the 95% confidence interval
t_critical = t.ppf(1 - alpha/2, df)

# Step 7: Print the results
print("Sample Mean:", x_bar)
print("Standard Error:", round(se, 2))
print("t-statistic:", round(t_stat, 2))
print("Degrees of Freedom:", df)
print("Critical t-value:", round(t_critical, 2))
print("p-value (two-tailed):", round(p_val_two_tail, 3))

# Step 8: Plot the t-distribution with rejection regions
x = np.linspace(-4, 4, 400)
y = t.pdf(x, df)

plt.figure(figsize=(10,6))
plt.plot(x, y, label='t-distribution', color='black')

# Plot t-statistic line
plt.axvline(t_stat, color='blue', linestyle='--', label=f't-stat = {t_stat:.2f}')

# Plot critical values
plt.axvline(-t_critical, color='red', linestyle='--', label='Critical values')
plt.axvline(t_critical, color='red', linestyle='--')

# Shade rejection regions
plt.fill_between(x, y, where=(x < -t_critical) | (x > t_critical), color='red', alpha=0.2, label='Rejection Region')

plt.title("Two-Tailed t-Test Visualization")
plt.xlabel("t-values")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
